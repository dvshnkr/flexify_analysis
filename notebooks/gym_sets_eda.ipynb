```python
import os
import pandas as pd
from pathlib import Path
from datetime import datetime
```

```python
cwd = os.getcwd()
data_dir = Path(cwd).parent / 'data'
gym_sets_path = data_dir / 'fetched' / 'gym_sets.csv'

gym_sets = pd.read_csv(gym_sets_path)
gym_sets.head()
```

```python
gym_sets['datetime'] = gym_sets['created'].apply(lambda x: datetime.fromtimestamp(x))
gym_sets['date'] = gym_sets['datetime'].dt.date

gym_sets.head()
```

```python
from dotenv import load_dotenv
load_dotenv()

GYM_START_DATE = os.getenv('GYM_START_DATE')
GYM_START_DATE
```

```python
GYM_START_DATE = datetime.strptime(GYM_START_DATE, '%Y-%m-%d').date()
GYM_START_DATE
```

```python
gym_sets[gym_sets['date'] > GYM_START_DATE]
```
